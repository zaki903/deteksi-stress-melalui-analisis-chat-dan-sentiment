## **Balancing Data Label**

```python
# Mapping label positif dari df2
positive_labels_df2 = ['happy', 'love']
```

```python
# Filter data positif dari df2
df2_positive = df2[df2['label'].isin(positive_labels_df2)].copy()
df2_positive['Label'] = 'Positive'
df2_positive.rename(columns={'tweet': 'Text'}, inplace=True)

df1_renamed = df1.rename(columns={'Tweet': 'Text'})
```

```python
jumlah_negative = (df1_renamed['Label'] == 'Negative').sum()
jumlah_positive = (df1_renamed['Label'] == 'Positive').sum()
selisih = jumlah_negative - jumlah_positive

# Sampling data tambahan dari df2_positive
jumlah_df2_positive = len(df2_positive)
if selisih > 0 and jumlah_df2_positive >= selisih:
    df2_positive_sampled = df2_positive.sample(n=selisih, random_state=42)
elif selisih > 0:
    df2_positive_sampled = df2_positive
    print(f"Hanya tersedia {jumlah_df2_positive} data positif dari df2, tidak cukup untuk menyetarakan total.")
else:
    df2_positive_sampled = pd.DataFrame(columns=['Text', 'Label'])
```

```python
# Gabungkan dan reset index
df_final = pd.concat([df1_renamed, df2_positive_sampled], ignore_index=True)
df_final = df_final[['Text', 'Label']]
df_final = df_final.sample(frac=1, random_state=42).reset_index(drop=True)

df_final.head()
```

```python
df_final.info()

print("\nJumlah data:", len(df_final))
print("Label unik:", df_final['Label'].unique())
print("\nJumlah per Label:")
print(df_final['Label'].value_counts())

print("\nMissing values:\n", df_final.isnull().sum())

print("\nData Duplikat:\n", df_final.duplicated().sum())
```

```python
df_final.to_csv('stress_dataset.csv', index=False)
```
